### Verse 16 last two words of the verse go with the first strongs of the next verse. Need to add it back to the last strongs of its verse

Matthew 1

```python
    #for chapter in range(1,29):
    chapter = 3
    url = 'https://www.blueletterbible.org/nasb/mat/{}/1/p1/ss1/s_1001'.format(chapter)
    page = simple_get(url)
    html = BeautifulSoup(page, 'html.parser')
    paragraph = html.find_all('div',{'id':"bibleTable"})
    strongs = []
    texts = []
    vnum = []
    verse = ''
    for div in paragraph[0].find_all('div',{'class': re.compile("(nasb_)")}):
        if div['class'][0] not in ['nasb_CN','nasb_SH']:
            skip = 0
            strongs_counts = 0
            for a in div.findAll('span',{'class': 'yhwh'}):
                a.replaceWith('ORD')
            for match in div.findAll('span',{'class': 'nasb_smallcaps'}):
                match.replaceWithChildren()

            for match in div.findAll('em'):
                match.replaceWithChildren()
            for strnum in div.find_all('sup'):
                if skip ==0:
                    if strnum.next_sibling:
                        if strnum.next_sibling.string.strip() == '' and strnum.next_sibling.next_sibling and (re.search('H[0-9]', strnum.next_sibling.next_sibling.string) or re.search('G[0-9]', strnum.next_sibling.next_sibling.string)):
                            dos_strings = strnum.text# + ',' + strnum.next_sibling.next_sibling.string
                            strongs_counts+=1
                            while strnum.next_sibling and strnum.next_sibling.string.strip() == '' and strnum.next_sibling.next_sibling and (re.search('H[0-9]', strnum.next_sibling.next_sibling.string) or re.search('G[0-9]', strnum.next_sibling.next_sibling.string)):
                                dos_strings = dos_strings+ ',' + strnum.next_sibling.next_sibling.string
                                strnum = strnum.next_sibling.next_sibling
                                skip +=1
                            strongs.append(dos_strings)
                        else:
                            strongs.append(strnum.text)
                            strongs_counts+=1
                    else:
                        strongs.append(strnum.text)
                        strongs_counts+=1
                else:
                    skip = skip-1
            # for sup in div('sup'):
                #sup.decompose()
            for span in div('span'):
                if span['class'][0] != 'vnum':
                    span.decompose()
            for a in div('a'):
                a.decompose()
            textspm = []
            text = ''
            prev_child = ''
            for child in list(div.children):
                if child != None and child.text.strip() !='':
                    print(child)
                    if child.name == None:
                        text+= child
                    elif child.name == 'span' and child['class'][0] == 'vnum':
                        text+= '--- verse_num '+child.text.split('\xa0')[0]+' ---'
                        #if prev_child.name == None and prev_child.name != 'sup':
            for words in text.split('---'):
                if len(words.strip())>0:
                    if words.find('verse_num')>-1:
                        verse = words.replace('verse_num','').strip()
                    else:
                        for t in words.split('\xa0'):
                            if len(re.sub(r'[^a-zA-Z0-9 ]', '', t).strip().lower())>0:
                                textspm.append(t.strip())
                                vnum.append(verse)
            textspm = [re.sub(r'[^a-zA-Z0-9 ]', '', t).strip().lower() for t in textspm if re.sub(r'[^a-zA-Z0-9 ]', '', t).strip().lower() != '']
            if len(textspm) == strongs_counts +1:
                textspm[len(textspm)-2] = textspm[len(textspm)-2]+ " "+textspm[len(textspm)-1]
                textspm = textspm[:len(textspm)-1]
                vnum = vnum[:len(vnum)-1]
            texts = texts + textspm   
    print(chapter, len(texts) , len(strongs), len(vnum))
```
